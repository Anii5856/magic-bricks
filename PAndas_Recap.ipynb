```python
# How to Treat Null Values.
# There are 3 Methods to treat Null Values.

# 1. Bfill and Ffill
# 2. Fill NA (Custom Fill -->> MEan,MEdian)
# 3. Drop Na (It will drop all the Null values available in the Dataset.)
```

```python
# 1. Bfill and Ffill

# NOTE : Try to work with the specific column when filling missing values. (For MEthod 1 & 2).

# How to check the index where the null values are available.

df[df.Parking.isnull()]
```

```python
# 1. Bfill and Ffill

# NOTE : Try to work with the specific column when filling missing values. (For MEthod 1 & 2).


df.Bathroom = df.Bathroom.ffill()
```

```python
df.isnull().sum()
```

```python
df.Parking.median()
```

```python
# Method 2 ()Fill NA fill with median is the best option.

df.Parking = df.Parking.fillna(df.Parking.median())
```

```python
df.isna().sum()
```

```python
df.Per_Sqft = df.Per_Sqft.fillna(df.Per_Sqft.median())
```

```python
df.isnull().sum()
```

```python
# Method 3: Drop Na (Drop NA won't work with Specific Column it will work with entire dataset.)

df = df.dropna()
```
